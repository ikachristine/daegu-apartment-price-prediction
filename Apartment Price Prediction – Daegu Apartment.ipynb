## **Business Problem Understanding**

### **Background**

Apartemen merupakan salah satu solusi utama dalam memenuhi kebutuhan hunian masyarakat modern, khususnya di wilayah perkotaan dengan keterbatasan lahan dan tingginya aktivitas ekonomi. Pertumbuhan populasi dan urbanisasi yang pesat menyebabkan permintaan terhadap hunian vertikal terus meningkat, menjadikan apartemen sebagai pilihan yang efisien dan strategis.

Harga apartemen dipengaruhi oleh berbagai faktor, baik internal maupun eksternal. Faktor internal meliputi karakteristik fisik seperti luas unit, tahun pembangunan, dan fasilitas gedung, sedangkan faktor eksternal mencakup lokasi, akses transportasi, serta ketersediaan fasilitas publik di sekitar properti. Kombinasi faktor-faktor tersebut membentuk nilai suatu unit di pasar.

### **Problem Statement**

Dalam praktiknya, penentuan harga apartemen sering dilakukan tanpa analisis data yang komprehensif. Hal ini berpotensi menyebabkan ketidaksesuaian dengan harga pasar.

Penetapan harga yang terlalu tinggi dapat memperpanjang waktu penjualan (time on market), sedangkan harga yang terlalu rendah dapat menyebabkan kehilangan potensi keuntungan. Oleh karena itu, diperlukan pendekatan berbasis data untuk membantu menentukan harga yang lebih akurat dan optimal.

### **Objective**

Membangun model prediksi harga apartemen menggunakan teknik machine learning untuk membantu pemilik, agen properti, dan investor dalam menetapkan harga yang lebih akurat, kompetitif, dan berbasis data.

### **Stakeholder**

- **Real Estate Agent**  
  Menggunakan model untuk menentukan harga listing yang kompetitif agar properti lebih cepat terjual tanpa kehilangan potensi keuntungan.

- **Property Owner (Individual / Developer)**  
  Menggunakan model sebagai referensi dalam menetapkan harga jual yang optimal sesuai kondisi pasar.

- **Buyer / Investor**  
  Menggunakan hasil prediksi untuk mengevaluasi apakah harga properti wajar dan mendukung pengambilan keputusan pembelian atau investasi.

### **Goals**

Berdasarkan permasalahan yang telah diidentifikasi, proyek ini bertujuan untuk mengembangkan model prediksi harga apartemen berbasis machine learning yang mampu mengestimasi harga secara objektif dan berbasis data historis.

Model ini diharapkan dapat berfungsi sebagai decision-support tool bagi berbagai stakeholder, seperti pemilik properti (individual owner/developer), agen properti, serta investor, dalam menetapkan harga jual yang lebih akurat dan kompetitif sesuai dengan karakteristik properti dan kondisi pasar.

Dengan memanfaatkan data historis, model akan mempelajari hubungan antara berbagai faktor properti dan nilai pasar apartemen. Pendekatan ini memungkinkan proses penentuan harga tidak lagi hanya bergantung pada intuisi atau perbandingan sederhana, tetapi didasarkan pada analisis kuantitatif yang terukur.

Melalui penggunaan model ini, diharapkan risiko overpricing (harga terlalu tinggi) dan underpricing (harga terlalu rendah) dapat diminimalkan, serta mendukung pengambilan keputusan yang lebih tepat, strategis, dan berbasis data.

Model ini dirancang untuk membantu perusahaan dalam menyediakan estimasi harga yang objektif dan berbasis data, sehingga dapat digunakan oleh pemilik individu maupun developer dalam menentukan harga jual yang lebih tepat dan kompetitif. 

### **Metric Evaluation**

Evaluasi metrik yang digunakan dalam model ini adalah RMSE, MAE, dan MAPE. RMSE merupakan nilai akar dari rata-rata kuadrat error sehingga tetap memberikan penalti lebih besar terhadap kesalahan yang ekstrem, namun tetap berada dalam satuan harga asli sehingga lebih mudah diinterpretasikan. MAE adalah rata-rata nilai absolut dari error yang menggambarkan besar rata-rata selisih prediksi terhadap nilai aktual secara langsung. Sementara itu, MAPE menunjukkan rata-rata persentase kesalahan prediksi, sehingga lebih mudah dipahami dalam konteks bisnis karena memberikan gambaran seberapa besar model meleset secara proporsional terhadap harga sebenarnya.

MSE tidak digunakan sebagai metrik utama karena nilainya berada dalam satuan kuadrat (won²) sehingga sulit untuk diinterpretasikan secara praktis dalam konteks harga properti. Selain itu, R-square (R²) tidak dipilih sebagai metrik utama karena hanya menunjukkan seberapa besar variasi target yang dapat dijelaskan oleh model, tetapi tidak memberikan informasi mengenai besarnya kesalahan prediksi dalam satuan harga aktual. Dalam konteks estimasi harga apartemen, metrik yang menunjukkan besar error secara langsung (RMSE dan MAE) serta dalam bentuk persentase (MAPE) dinilai lebih relevan dan informatif untuk pengambilan keputusan.

Semakin kecil nilai RMSE, MAE, dan MAPE yang dihasilkan, maka model dapat dikatakan semakin akurat dalam memprediksi harga jual apartemen sesuai dengan keterbatasan fitur yang tersedia dalam dataset.

### **Outliers Detection**

Banyaknya outliers pada data tentu akan memengaruhi hasil analisis dan pemodelan yang dilakukan. Ada beberapa hal yang harus dipertimbangkan sebelum memutuskan treatment yang sesuai dengan data, di antaranya adalah dengan mempertimbangkan sisi domain knowledge.

```python
# Plot boxplot untuk mengecek outliers
df_model.boxplot(rot=90); 
```

Berdasarkan boxplot, terlihat adanya outlier yang cukup signifikan pada fitur HargaJual dan LuasApartemen. Pada fitur lain kemungkinan besar outlier tidak terlihat karena skala grafik tertarik oleh HargaJual. Dalam proyek ini, untuk menghasilkan dataset yang lebih representatif dan stabil, akan dilakukan penanganan outlier dengan mempertimbangkan kombinasi antara metode statistik (IQR) dan pertimbangan domain knowledge.

Dengan demikian, model yang dibangun nantinya akan lebih optimal dalam memprediksi harga apartemen pada rentang nilai yang terdapat dalam dataset yang telah dibersihkan (clean dataset). Namun, perlu dicatat bahwa kemampuan prediksi model akan terbatas pada rentang tersebut, sehingga prediksi untuk nilai di luar distribusi data pelatihan berpotensi menghasilkan bias atau akurasi yang menurun.

```python
# Membuat fungsi untuk mengecek outliers
def find_anomalies(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    limit = iqr*1.5
    print(f'IQR: {iqr}')

    limit_bawah = q1 - limit
    limit_atas = q3 + limit
    print(f'limit_bawah: {limit_bawah}')
    print(f'limit_atas: {limit_atas}')        
```

```python
# Outliers pada fitur HargaJual
find_anomalies(df_model['HargaJual'])
```

```python
# Outliers pada fitur LuasApartemen
find_anomalies(df_model['LuasApartemen'])
```

```python
# Outliers pada fitur TahunDibangun
find_anomalies(df_model['TahunDibangun'])
```

```python
# Outliers pada fitur JumlahFasilitasDalamApartemen
find_anomalies(df_model['JumlahFasilitasDalamApartemen'])
```

```python
# Outliers pada fitur JumlahParkirBasement
find_anomalies(df_model['JumlahParkirBasement'])
```

## **Modeling**

```python
# Import library untuk modeling

from sklearn.model_selection import train_test_split, cross_val_score, RandomizedSearchCV, GridSearchCV, KFold

import category_encoders as ce
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost.sklearn import XGBRegressor
from sklearn.compose import TransformedTargetRegressor

from sklearn.preprocessing import StandardScaler

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
```

### **Encoding**

Agar dapat menggunakan semua fitur yang dipilih, maka fitur-fitur kategorikal harus diubah menjadi fitur numerikal. Kita dapat menggunakan ColumnTransformer untuk dapat mengubah tipe datanya, yang mana nanti dapat diaplikasikan pada pipeline model.

```python
# Mengubah fitur kategorikal menjadi numerikal

ordinal_order = [
    ['0-5min',
     '5min~10min',
     '10min~15min',
     '15min~20min',
     'no_bus_stop_nearby']
]

transformer = ColumnTransformer([
    ('onehot', OneHotEncoder(drop='first'), ['TipeApartemen', 'StasiunSubway']),
    ('ordinal', OrdinalEncoder(categories=ordinal_order), ['WaktuKeSubway'])
], remainder='passthrough')
```

**One-Hot Encoding** digunakan pada fitur kategorikal yang tidak memiliki urutan (nominal), seperti lokasi atau kategori tertentu. Metode ini mengubah setiap kategori menjadi kolom biner (0 dan 1), sehingga model tidak menganggap adanya hubungan urutan antar kategori. Hal ini penting untuk menghindari bias dalam interpretasi model.

**Ordinal Encoding** digunakan pada fitur kategorikal yang memiliki urutan (ordinal), seperti tingkat kualitas atau level tertentu. Dengan pendekatan ini, kategori diubah menjadi nilai numerik berdasarkan urutannya, sehingga informasi hierarki tetap terjaga dan dapat dimanfaatkan oleh model.

Penggunaan kedua metode ini bertujuan untuk memastikan bahwa data kategorikal dapat direpresentasikan secara numerik tanpa menghilangkan makna aslinya, sehingga model dapat belajar pola hubungan antara fitur dan harga apartemen dengan lebih optimal.

### **Data Splitting**

```python
# Memisahkan data independen variabel dengan target
X = df_model.drop(['HargaJual'], axis=1)
y = df_model['HargaJual']
```

### **Train and Test Splitting**

```python
# Splitting data training dan test dengan proporsi 70:30
X_train, X_test, y_train, y_test= train_test_split(
    X,
    y,  
    test_size=0.3,
    random_state=1)
```

### **Choose a Benchmark Model**

```python
# Define algoritma yang digunakan
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state=1)
rf = RandomForestRegressor(random_state=1)
xgb = XGBRegressor(random_state=1)

# Pemodelan dilakukan dalam skala logaritmik, namun kemudian di-inverse kembali untuk interpretasi
log_lr = TransformedTargetRegressor(lr, func=np.log, inverse_func=np.exp)
log_knn = TransformedTargetRegressor(knn, func=np.log, inverse_func=np.exp)
log_dt = TransformedTargetRegressor(dt, func=np.log, inverse_func=np.exp)
log_rf = TransformedTargetRegressor(rf, func=np.log, inverse_func=np.exp)
log_xgb = TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp)

# Scaling data
scaler = StandardScaler()

# Kandidat algoritma
models = [log_lr, log_knn, log_dt, log_rf, log_xgb]

score_rmse = []
nilai_mean_rmse = []
nilai_std_rmse = []

score_mae = []
nilai_mean_mae = []
nilai_std_mae = []

score_mape = []
nilai_mean_mape = []
nilai_std_mape = []

# Mencari algoritma terbaik berdasarkan tiap metrics
for i in models:
    
    crossval = KFold(n_splits=5, shuffle=True, random_state=1)

    estimator = Pipeline([
        ('preprocessing', transformer),
        ('scaler', scaler),
        ('model', i)
    ])

    # RMSE
    model_cv_rmse = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_root_mean_squared_error', 
        error_score='raise'
        )

    print(model_cv_rmse, i)

    score_rmse.append(model_cv_rmse)
    nilai_mean_rmse.append(model_cv_rmse.mean())
    nilai_std_rmse.append(model_cv_rmse.std())

    # MAE
    model_cv_mae = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_error', 
        error_score='raise'
        )

    print(model_cv_mae, i)

    score_mae.append(model_cv_mae)
    nilai_mean_mae.append(model_cv_mae.mean())
    nilai_std_mae.append(model_cv_mae.std())

    # MAPE
    model_cv_mape = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_percentage_error', 
        error_score='raise'
        )

    print(model_cv_mape, i)

    score_mape.append(model_cv_mape)
    nilai_mean_mape.append(model_cv_mape.mean())
    nilai_std_mape.append(model_cv_mape.std())
```

```python
# Evaluasi hasil dari 5 kandidat algoritma yang digunakan
pd.DataFrame({
    'Model': ['Linear Regression', 'KNN Regressor', 'DecisionTree Regressor', 'RandomForest Regressor', 'XGBoost Regressor'],
    'Mean_RMSE': nilai_mean_rmse,
    'Std_RMSE': nilai_std_rmse,
    'Mean_MAE': nilai_mean_mae,
    'Std_MAE': nilai_std_mae,
    'Mean_MAPE': nilai_mean_mape,
    'Std_MAPE': nilai_std_mape
})
```

Terdapat perbedaan cukup signifikan antara nilai RMSE dan MAE, di mana nilai RMSE lebih tinggi karena nilai residuals atau error-nya dikuadratkan terlebih dahulu sebelum dirata-ratakan. Hal ini menyebabkan RMSE akan memberikan 'weight' yang lebih tinggi untuk nilai error yang besar. Dengan kata lain, terdapat nilai-nilai error yang besar yang dihasilkan oleh semua algoritma yang digunakan, sehingga ada perbedaan signifikan antara nilai RMSE dan MAE.

- Berdasarkan nilai RMSE, XGBoost adalah model terbaik. Sedangkan jika berdasarkan nilai MAE dan MAPE, RandomForest memiliki nilai paling rendah.
- Selanjutnya, akan dilakukan prediksi pada test set dengan 2 benchmark model terbaik, yaitu XGBoost dan RandomForest. 

### **Predict to Test Set with the Benchmark Model**

Prediksi pada test set dengan menggunakan model XGBoost dan Random Forest

```python
# Benchmark 2 model terbaik
models = {
    'XGB': XGBRegressor(random_state=1),
    'RandomForest': RandomForestRegressor(random_state=1)
}

score_rmse = []
score_mae = []
score_mape = []

# Prediksi pada test set
for i in models:

    model = Pipeline([
        ('preprocessing', transformer),
        ('scaler', scaler),
        ('model', models[i])
        ])

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    score_rmse.append(np.sqrt(mean_squared_error(y_test, y_pred)))
    score_mae.append(mean_absolute_error(y_test, y_pred))
    score_mape.append(mean_absolute_percentage_error(y_test, y_pred))

score_before_tuning = pd.DataFrame({'RMSE': score_rmse, 'MAE': score_mae, 'MAPE': score_mape}, index=models.keys())
score_before_tuning
```

- Ketika dilakukan prediksi pada test set, performa RandomForest secara akumulatif sedikit lebih baik. Hal ini dapat dilihat dari nilai RMSE dan MAE yang sedikit lebih rendah dari XGBoost.


### **Hyperparameter Tuning**

Karena model RandomForest memiliki performa yang sedikit lebih baik dari XGBoost, maka selanjutnya kita akan menggunakan model RandomForest ini sebagai model akhir. Pada tahap ini, kita akan melakukan hyperparameter tuning pada model RandomForest dengan harapan dapat meningkatkan performa model.

```python
# Kedalaman pohon
max_depth = list(np.arange(1, 21))

# Jumlah pohon
n_estimators = list(np.arange(100, 501, 50))

# Minimum sample untuk split
min_samples_split = list(np.arange(2, 11))

# Minimum sample per leaf
min_samples_leaf = list(np.arange(1, 11))

# Jumlah fitur yang digunakan tiap pohon
max_features = ['sqrt', 'log2', None]

# Bootstrap
bootstrap = [True, False]

hyperparam_space_rf = {
    'model__n_estimators': n_estimators,
    'model__max_depth': max_depth,
    'model__min_samples_split': min_samples_split,
    'model__min_samples_leaf': min_samples_leaf,
    'model__max_features': max_features,
    'model__bootstrap': bootstrap
}
```

```python
# Inisialisasi model RandomForest
rf = RandomForestRegressor(random_state=1)

# Membuat pipeline
estimator_rf = Pipeline([
        ('preprocessing', transformer),
        ('scaler', scaler),   # boleh dihapus jika mau lebih efisien
        ('model', rf)
])

# Cross validation
crossval = KFold(n_splits=5, shuffle=True, random_state=1)

# Hyperparameter space untuk RandomForest
hyperparam_space_rf = {
    'model__n_estimators': np.arange(100, 501, 50),
    'model__max_depth': np.arange(5, 31, 5),
    'model__min_samples_split': np.arange(2, 11),
    'model__min_samples_leaf': np.arange(1, 11),
    'model__max_features': ['sqrt', 'log2', None],
    'model__bootstrap': [True, False]
}

# Hyperparameter tuning
random_rf = RandomizedSearchCV(
    estimator_rf, 
    param_distributions=hyperparam_space_rf,
    n_iter=50,
    cv=crossval,
    scoring=['neg_root_mean_squared_error', 
             'neg_mean_absolute_error', 
             'neg_mean_absolute_percentage_error'],
    n_jobs=-1,
    refit='neg_root_mean_squared_error',
    random_state=1
)
```

```python
# Fitting data training untuk mencari parameter terbaik
random_rf.fit(X_train, y_train)
```

```python
# Melihat hasil tuning dalam bentuk dataframe. Diurutkan berdasarkan RMSE, MAE & MAPE
pd.DataFrame(random_rf.cv_results_).sort_values(by=['rank_test_neg_root_mean_squared_error', 'rank_test_neg_mean_absolute_error', 'rank_test_neg_mean_absolute_percentage_error']).head()
```

```python
# Cek skor dan params terbaik
print('RandomForest')
print('Best_score:', random_rf.best_score_)
print('Best_params:', random_rf.best_params_)
```

### **Predict to Test Set with the Tuned Model**

Melakukan prediksi pada test set dengan menggunakan model RandomForest dan hyperparameter terpilih.

```python
# Model RandomForest
model = {'RandomForest': RandomForestRegressor(random_state=1)}

# Ambil estimator terbaik hasil tuning
rf_tuning = random_rf.best_estimator_

# Fitting model
rf_tuning.fit(X_train, y_train)

# Predict test set
y_pred_rf_tuning = rf_tuning.predict(X_test)

# Hitung metrics setelah tuning
rmse_rf_tuning = np.sqrt(mean_squared_error(y_test, y_pred_rf_tuning))
mae_rf_tuning = mean_absolute_error(y_test, y_pred_rf_tuning)
mape_rf_tuning = mean_absolute_percentage_error(y_test, y_pred_rf_tuning)

# Simpan ke dataframe
score_after_tuning = pd.DataFrame({
    'RMSE': rmse_rf_tuning,
    'MAE': mae_rf_tuning,
    'MAPE': mape_rf_tuning
}, index=model.keys())

score_after_tuning
```
